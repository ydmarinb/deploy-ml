En esta sección el proposito es automatizar el pipeline del machine learning:

> Extracción de de base de datos ->Analisis de datos -> ingenieria de caracteristicas -> **Selección de caracteristicas** -> Creación del modelo -> Predicción

En el aquellos proceso donde se tenga dependencia de la aletoriaedad es indispensable definir una semilla para garantizar la reproducibilidad

```python
# manejo de bases de datos
import pandas as pd
import numpy as np

# creación de graficas
import matplotlib.pyplot as plt
import seaborn as sns

# Construir modelo
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# Mostrar todas las columnas en el pandas data frame
pd.pandas.set_option('display.max_columns', None)
```

```python
# Cargar los conjuntos de entrenamiento y test

X_train = pd.read_csv('xtrain.csv')
X_test = pd.read_csv('xtest.csv')

X_train.head()
```

```python
# Cargar las variables Y de entrenamiento y test
y_train = pd.read_csv('ytrain.csv')
y_test = pd.read_csv('ytest.csv')

y_train.head()
```

# Ingenieria de caracteristicas

```python

sel_ = SelectFromModel(Lasso(alpha=0.001, random_state=0))

# Se entrena el modelo laso para la seleccion de variables
sel_.fit(X_train, y_train)
```

```python
# Observemos las variables que se seleccionaron(True)
sel_.get_support()
```

```python
# Filtramos las variables seleccionadas
selected_feats = X_train.columns[(sel_.get_support())]
```

```python
# Guardamos data set con las varianles seleccionadas
pd.Series(selected_feats).to_csv('selected_features.csv', index=False)
```
