```python
# Manejar bases de datos
import pandas as pd
import numpy as np

# para graficas
import matplotlib.pyplot as plt

# Guardar los modelos
import joblib

# Cosntruir el modelo
from sklearn.linear_model import Lasso

# Evaluar los modelo
from sklearn.metrics import mean_squared_error, r2_score

# Visualizar todas las columnas de df
pd.pandas.set_option('display.max_columns', None)
```

```python
# Cargar los datos de train

X_train = pd.read_csv('xtrain.csv')
X_test = pd.read_csv('xtest.csv')

X_train.head()
```

```python
# Cargar los datos de test
y_train = pd.read_csv('ytrain.csv')
y_test = pd.read_csv('ytest.csv')

y_train.head()
```

```python
# Cargar las categorias preseleccionadas

features = pd.read_csv('selected_features.csv')
features = features['0'].to_list() 

# display final feature set
features
```

```python
# Seleccionar estas variables en test y training
X_train = X_train[features]
X_test = X_test[features]
```

# Entrenar modelos

```python


lin_model = Lasso(alpha=0.001, random_state=0)

lin_model.fit(X_train, y_train)
```

```python
# Evaluar el modelo


# hacer prediccionar con el train
pred = lin_model.predict(X_train)

# r2, mse, rmse
print('train mse: {}'.format(int(
    mean_squared_error(np.exp(y_train), np.exp(pred)))))
print('train rmse: {}'.format(int(
    mean_squared_error(np.exp(y_train), np.exp(pred), squared=False))))
print('train r2: {}'.format(
    r2_score(np.exp(y_train), np.exp(pred))))
print()

# Hacer predicciones para test
pred = lin_model.predict(X_test)

# determine mse, rmse and r2
print('test mse: {}'.format(int(
    mean_squared_error(np.exp(y_test), np.exp(pred)))))
print('test rmse: {}'.format(int(
    mean_squared_error(np.exp(y_test), np.exp(pred), squared=False))))
print('test r2: {}'.format(
    r2_score(np.exp(y_test), np.exp(pred))))
print()

print('Average house price: ', int(np.exp(y_train).median()))
```

```python
# Evaluamos la predicciones
plt.scatter(y_test, lin_model.predict(X_test))
plt.xlabel('True House Price')
plt.ylabel('Predicted House Price')
plt.title('Evaluation of Lasso Predictions')
```

# Evaluamos la distribución de lso errores

```python
y_test.reset_index(drop=True, inplace=True)

preds = pd.Series(lin_model.predict(X_test))
```

```python
# Se puede ver un comportamiento semi normal

errors = y_test['SalePrice'] - preds
errors.hist(bins=30)
plt.show()
```

# Guardar el modelo


```python
joblib.dump(lin_model, 'linear_regression.joblib') 
```
