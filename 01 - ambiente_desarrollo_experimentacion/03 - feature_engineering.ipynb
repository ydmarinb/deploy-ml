En esta sección el proposito es automatizar el pipeline del machine learning:

> Extracción de de base de datos ->Analisis de datos -> **ingenieria de caracteristicas** -> Selección de caracteristicas -> Creación del modelo -> Predicción

En el aquellos proceso donde se tenga dependencia de la aletoriaedad es indispensable definir una semilla para garantizar la reproducibilidad

```python
# manejo de bases de datos
import pandas as pd
import numpy as np

# creación de graficas
import matplotlib.pyplot as plt
import seaborn as sns

# analisis estadisitico
import scipy.stats as stats

# Dividir datos en entrenamiento y text
from sklearn.model_selection import train_test_split

# Escalado de caracteristicas
from sklearn.preprocessing import MinMaxScaler

# para guardar objetos
import joblib

# Mostrar todas las columnas en el pandas data frame
pd.pandas.set_option('display.max_columns', None)
```

```python
# Cargar conjunto de datos
data = pd.read_csv('./data/train.csv')

# Dimensiones de la base de datos
print(data.shape)

# Ver el conjunto de datos
data.head()
```

# Cargar conjunto de datos entre entrenameinto y test

```python
# Como este proceso es aletorio se define una semilla

X_train, X_test, y_train, y_test = train_test_split(
    data.drop(['Id', 'SalePrice'], axis=1), 
    data['SalePrice'], # target
    test_size=0.1, 
    random_state=0,
)

X_train.shape, X_test.shape
```

# Ingenieria de caracteristicas

* Valores vacios
* Variables temporales
* Variables con distribuciones no normal
* Removar categorias raras
* Convertir string a numero
* Escalar las varianles

### Varianbles Y

Se aplica transformación logaritmica

```python
y_train = np.log(y_train)
y_test = np.log(y_test)
```

```python
# Ver notebook 02 para explicacion

cat_vars = [var for var in data.columns if data[var].dtype == 'O']


cat_vars = cat_vars + ['MSSubClass']


X_train[cat_vars] = X_train[cat_vars].astype('O')
X_test[cat_vars] = X_test[cat_vars].astype('O')
```

```python


cat_vars_with_na = [
    var for var in cat_vars
    if X_train[var].isnull().sum() > 0
]


X_train[cat_vars_with_na ].isnull().mean().sort_values(ascending=False)
```

```python
# Se calculan aquellas variables que tienen mas de 10% de datos vacios
with_string_missing = [
    var for var in cat_vars_with_na if X_train[var].isnull().mean() > 0.1]

# Se calculan aquellas variables que tienen mes de 10% de datos vacios
with_frequent_category = [
    var for var in cat_vars_with_na if X_train[var].isnull().mean() < 0.1]
```

```python
# Remplazar valores varios con : "Missing"

X_train[with_string_missing] = X_train[with_string_missing].fillna('Missing')
X_test[with_string_missing] = X_test[with_string_missing].fillna('Missing')
```

```python
# Reemplazar las variabels viables a imputar con la moda(<10)
for var in with_frequent_category:
    
    mode = X_train[var].mode()[0]
    
    print(var, mode)
    
    X_train[var].fillna(mode, inplace=True)
    X_test[var].fillna(mode, inplace=True)
```

```python
# Evaluación de los missing despues de aplicar el proceso

X_train[cat_vars_with_na].isnull().sum()
```

### Numericas variables

Se recurre a realizar la imputación con la media. Esta es un ejemplo esquematico porque en lo general no es la mejor opción reemplazar por media.

```python

num_vars = [
    var for var in X_train.columns if var not in cat_vars and var != 'SalePrice'
]
```

```python
# Lista de variables numericas que tienen valores vacios
vars_with_na = [
    var for var in num_vars
    if X_train[var].isnull().sum() > 0
]

# Imprimir el porcentaje missing por variable
X_train[vars_with_na].isnull().mean()
```

```python
# Se procede a la imputación

for var in vars_with_na:

    # Calcular la media usando el conjunto de entrenamiento
    mean_val = X_train[var].mean()
    
    print(var, mean_val)

    # Se agregan variables indicadores de missing values
    X_train[var + '_na'] = np.where(X_train[var].isnull(), 1, 0)
    X_test[var + '_na'] = np.where(X_test[var].isnull(), 1, 0)

    # reemplazar la missing values con la media
    X_train[var].fillna(mean_val, inplace=True)
    X_test[var].fillna(mean_val, inplace=True)

# Chequear que las varible quedo sin nulos despues de la impitación
X_train[vars_with_na].isnull().sum()
```

### Variables temporales
En el previo notebook vimos que hay 4 variables temporales. Tambien se hicieron transformaciones para calcular la diferencia entre las variables y el año que se vendio.

```python
def elapsed_years(df, var):
    # Tomas la diferencia entre el año de venta y las
    # demas variables
    df[var] = df['YrSold'] - df[var]
    return df
```

```python

for var in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    X_train = elapsed_years(X_train, var)
    X_test = elapsed_years(X_test, var)
```

```python
# Se borra la variable de respuesta de entrenamiento y prueba
X_train.drop(['YrSold'], axis=1, inplace=True)
X_test.drop(['YrSold'], axis=1, inplace=True)
```

### Transformación de variables numericas
Transformamos con log algunas variabels que vimos anteriormente donde este tipo de transformación funcionaba bien.

```python
for var in ["LotFrontage", "1stFlrSF", "GrLivArea"]:
    X_train[var] = np.log(X_train[var])
    X_test[var] = np.log(X_test[var])
```

Tambien aplicamos las transformación Yeo-jhnson para la variable en la que funciono bien.

```python
 X_train['LotArea'], param = stats.yeojohnson(X_train['LotArea'])

X_test['LotArea'] = stats.yeojohnson(X_test['LotArea'], lmbda=param)

print(param)
```

Ahora aplicamos un proceos de binarización a las variables altamente sesgadas

```python
skewed = [
    'BsmtFinSF2', 'LowQualFinSF', 'EnclosedPorch',
    '3SsnPorch', 'ScreenPorch', 'MiscVal'
]

for var in skewed:
    
    X_train[var] = np.where(X_train[var]==0, 0, 1)
    X_test[var] = np.where(X_test[var]==0, 0, 1)
```

# Transformar variables categoricas

A aquellas variables categoricas que tiene un orden implicito se realiza una recodificación.


```python


qual_mappings = {'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5, 'Missing': 0, 'NA': 0}

qual_vars = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond',
             'HeatingQC', 'KitchenQual', 'FireplaceQu',
             'GarageQual', 'GarageCond',
            ]

for var in qual_vars:
    X_train[var] = X_train[var].map(qual_mappings)
    X_test[var] = X_test[var].map(qual_mappings)
```

```python
exposure_mappings = {'No': 1, 'Mn': 2, 'Av': 3, 'Gd': 4}

var = 'BsmtExposure'

X_train[var] = X_train[var].map(exposure_mappings)
X_test[var] = X_test[var].map(exposure_mappings)
```

```python
finish_mappings = {'Missing': 0, 'NA': 0, 'Unf': 1, 'LwQ': 2, 'Rec': 3, 'BLQ': 4, 'ALQ': 5, 'GLQ': 6}

finish_vars = ['BsmtFinType1', 'BsmtFinType2']

for var in finish_vars:
    X_train[var] = X_train[var].map(finish_mappings)
    X_test[var] = X_test[var].map(finish_mappings)
```

```python
garage_mappings = {'Missing': 0, 'NA': 0, 'Unf': 1, 'RFn': 2, 'Fin': 3}

var = 'GarageFinish'

X_train[var] = X_train[var].map(garage_mappings)
X_test[var] = X_test[var].map(garage_mappings)
```

```python
qual_vars = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond',
             'HeatingQC', 'KitchenQual', 'FireplaceQu',
             'GarageQual', 'GarageCond',
            ]

finish_vars = ['BsmtFinType1', 'BsmtFinType2']
```

```python
fence_mappings = {'Missing': 0, 'NA': 0, 'MnWw': 1, 'GdWo': 2, 'MnPrv': 3, 'GdPrv': 4}

var = 'Fence'

X_train[var] = X_train[var].map(fence_mappings)
X_test[var] = X_test[var].map(fence_mappings)
```

A continuación, todas aquellas cartegorias que representa menos del 10% de las observaciones seran agrupadas en una nueva categoria.

```python

qual_vars  = qual_vars + finish_vars + ['BsmtExposure','GarageFinish','Fence']



cat_others = [
    var for var in cat_vars if var not in qual_vars
]

len(cat_others)
```

```python
def find_frequent_labels(df, var, rare_perc):
    
    # funcion para buscar categorias representas en menos derare_perc

    df = df.copy()

    tmp = df.groupby(var)[var].count() / len(df)

    return tmp[tmp > rare_perc].index


for var in cat_others:
    
    
    frequent_ls = find_frequent_labels(X_train, var, 0.01)
    
    print(var, frequent_ls)
    print()
    
    # Se reemplazan categorias raras con "Rare"
    X_train[var] = np.where(X_train[var].isin(
        frequent_ls), X_train[var], 'Rare')
    
    X_test[var] = np.where(X_test[var].isin(
        frequent_ls), X_test[var], 'Rare')
```

Vamos a codificar las variables categoricas como numeros

```python
# Esta función asigna un valores menor numerico a las categorias con menos precio de hpuse price

def replace_categories(train, test, y_train, var, target):
    
    tmp = pd.concat([X_train, y_train], axis=1)
    
    # Ordenar variables con base en el valor de su media
    ordered_labels = tmp.groupby([var])[target].mean().sort_values().index

    # Crear diccionario para recodificar
    ordinal_label = {k: i for i, k in enumerate(ordered_labels, 0)}
    
    print(var, ordinal_label)
    print()

    # usa diccionario para transformar las variables
    train[var] = train[var].map(ordinal_label)
    test[var] = test[var].map(ordinal_label)
```

```python
for var in cat_others:
    replace_categories(X_train, X_test, y_train, var, 'SalePrice')
```

```python
# visualicemos

def analyse_vars(train, y_train, var):
    
    # function plots median house sale price per encoded
    # category
    
    tmp = pd.concat([X_train, np.log(y_train)], axis=1)
    
    tmp.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.ylim(2.2, 2.6)
    plt.ylabel('SalePrice')
    plt.show()
    
for var in cat_others:
    analyse_vars(X_train, y_train, var)
```

# Escalado de caracteristicas
Para construir modelos lineales es importante escalar las avriables

```python

# Scalador max-min
scaler = MinMaxScaler()

#  Ajustar el salcer al conjubto de entrenamiento
scaler.fit(X_train) 



X_train = pd.DataFrame(
    scaler.transform(X_train),
    columns=X_train.columns
)

X_test = pd.DataFrame(
    scaler.transform(X_test),
    columns=X_train.columns
)
```
